```python
diabetesresults=pd.read_csv('diabetes_binary_5050split_health_indicators_BRFSS2015_cleaned.csv')
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets=apriori(diabetesresults,min_support=0.35, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets, metric='lift', min_threshold=1)
```

```python
rules
```

```python
ant_rules = rules[rules['consequents'] == {'Diabetes_binary'}]
consq_rules = rules[rules['antecedents'] == {'Diabetes_binary'}]
```

```python
pd.concat([ant_rules,consq_rules])
```
